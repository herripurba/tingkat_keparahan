```python
ct =CRFTagger()
ct.set_model_file('all_indo_man_tag_corpus_model.crf.tagger')
def preprocessing(berita):
    s = berita
    s = s.replace('\n', ' ')
    s = s.replace('\r', ' ')
    s=re.sub("\(.*?\)", "", s)
    # s = s.lower()
    tokens = word_tokenize(s)
    return tokens
```

```python
def indeksKe(arr,indikator):
    for x in range(0, len(arr[0])):
        for y in indikator:
            y=y.split()
            if(len(y)==1 and arr[0][x][0]==y[0]):
                return x
            elif(len(y)==2 and len(arr[0])>x and arr[0][x][0]==y[0] and arr[0][x+1][0]==y[1]):
                return x
    return -1
```

```python
def matiScale(matiCount):
    matiCount=float(matiCount)
    if matiCount ==0 :
      return 1
    elif matiCount > 0 and matiCount <= 3:
      return 2
    return 3

def lukaScale(lukaCount):
    lukaCount=float(lukaCount)
    if lukaCount <= 5:
      return 1
    elif lukaCount >= 6 and lukaCount <= 10:
      return 2
    return 3

def hilangScale(hilangCount):
    if hilangCount < 4:
      return 1
    elif hilangCount >= 4 and hilangCount <= 10:
      return 2
    return 3

def tenggelamScale(tenggelamCount):
    if tenggelamCount < 4:
      return 1
    elif tenggelamCount >= 4 and tenggelamCount <= 10:
      return 2
    return 3

def pesawatScale(pesawatCount):
    if pesawatCount ==0:
      return 1
    elif pesawatCount ==1:
      return 3

def calculate(mati, luka, hilang, tenggelam,pesawat):
  subtractor = 0
  total = 0
  fiturCount = 5

  if (mati > 0):
    total += matiScale(matiCount=mati)
  else:
    subtractor += 1

  if (luka > 0):
    total += lukaScale(lukaCount=luka)
  else:
    subtractor += 1

  if (tenggelam > 0):
    total += tenggelamScale(tenggelamCount=tenggelam)
  else:
    subtractor += 1
  
  if (hilang > 0):
        total += hilangScale(hilangCount=hilang)
  else:
    subtractor += 1

  if (pesawat > 0):
    total += pesawatScale(pesawatCount=pesawat)
  else:
    subtractor += 1
          

  divider = (fiturCount - subtractor)
  severityMean = 0

  if (divider > 0):
    severityMean = total / divider

  if severityMean <= 1:
    return 'rendah'
  elif severityMean > 1 and severityMean <= 2:
    return 'sedang'
  else:
    return 'parah'
```
