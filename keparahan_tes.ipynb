```python
ct =CRFTagger()
ct.set_model_file('all_indo_man_tag_corpus_model.crf.tagger')
def preprocessing(berita):
    s = berita
    s = s.replace('\n', ' ')
    s = s.replace('\r', ' ')
    s=re.sub("\(.*?\)", "", s)
    # s = s.lower()
    tokens = word_tokenize(s)
    return tokens
```

```python
def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False
```

```python
def convertTeks(string):
    string = string.lower()
    string = string.split()
    
    arr = []

    if(len(string)==2 and isfloat(string[0])==True  and string[1]=='juta'):
        hasil="{:.0f}".format(float(string[0])*1000000)
        arr.append(hasil)
    elif(len(string)==2 and isfloat(string[0])==True  and string[1]=='ribu'):
        hasil="{:.0f}".format(float(string[0])*1000)
        arr.append(hasil)
    else:
        for teks in string:
            if(teks[-1:]=='%' and teks[:-1].isdigit()):
                arr.append(teks)
            elif(len(teks)==5 and teks[2]=='.' and isfloat(teks)==True):
                arr.append(str(1))
            elif(isfloat(teks)==True):
                arr.append(teks)
            elif(teks=="satu" or teks=="seorang" or teks=="seseorang" or teks=="sesosok"):
                arr.append(str(1))
            elif(teks=="dua" or teks=="kedua" or teks=="keduanya" or teks=="pasangan" or teks=='ke-2'):
                arr.append(str(2))
            elif(teks=="tiga" or teks=="ketiga" or teks=='ke-3'):
                arr.append(str(3))
            elif(teks=="empat" or teks=="keempat" or teks=='ke-4'):
                arr.append(str(4))
            elif(teks=="lima" or teks=="kelima" or teks=='ke-5'):
                arr.append(str(5))
            elif(teks=="enam"  or teks=="keenam" or teks=='ke-6'):
                arr.append(str(6))
            elif(teks=="tujuh"  or teks=="ketujuh" or teks=='ke-7'):
                arr.append(str(7))
            elif(teks=="delapan"  or teks=="kedelapan" or teks=='ke-8'):
                arr.append(str(8))
            elif(teks=="sembilan"  or teks=="kesembilan" or teks=='ke-9'):
                arr.append(str(9))
            elif(teks=="sepuluh"  or teks=="kesepuluh" or teks=="puluhan" or teks=='ke-10'):
                arr.append(str(10))
            elif(teks=="salah" or teks=="sebanyak"):
                arr.append('')
            elif(teks=="persen"):
                arr.append('%')
            elif(teks=="belasan"):
                arr.append(str(11))

            #Banyak yang di bawah ini tergantung topik masing2
            elif(teks=="sejumlah"):
                arr.append(str(random.randint(6,10)))
            elif(teks=="beberapa"):
                arr.append(str(random.randint(0,5)))
            elif(teks=="ratusan"):
                arr.append(str(100))
            elif(teks=="ribuan"):
                arr.append(str(1000))
            elif(teks=="ribu"):
                arr.append(str(000))
            elif(teks=="jutaan"):
                arr.append(str(1000000))
            elif(teks=="juta"):
                arr.append(str(000000))
            else:
                arr.append(teks)

    return("".join(arr))
```

```python
def indeksKe(arr,indikator):
    for x in range(0, len(arr[0])):
        for y in indikator:
            if(arr[0][x][0]==y):
                return x
    return -1
```
